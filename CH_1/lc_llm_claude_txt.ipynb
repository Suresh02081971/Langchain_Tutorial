```python
from langchain.chat_models import init_chat_model

# Initialize Claude model
model = init_chat_model(
    "claude-sonnet-4-5-20250929",  # Current Claude Sonnet 4.5 model
    temperature=0.5,
    timeout=10,
    max_tokens=1000
)

# Use the model
response = model.invoke("Hello, how are you?")
print(response)
```

```python
from langchain_anthropic import ChatAnthropic

# Initialize Claude Sonnet 4.5
llm = ChatAnthropic(
    model="claude-sonnet-4-5-20250929",
    temperature=0.7,
    max_tokens=1024
)

# Example 1: Simple question
print("Example 1: Simple Question")
print("-" * 50)
response = llm.invoke("What is Python?")
print(response.content)
print("\n")

# Example 2: Creative writing
print("Example 2: Creative Writing")
print("-" * 50)
response = llm.invoke("Write a haiku about coding")
print(response.content)
print("\n")

# Example 3: Technical explanation
print("Example 3: Technical Explanation")
print("-" * 50)
response = llm.invoke("Explain what an API is in one sentence")
print(response.content)
print("\n")

# Example 4: Code generation
print("Example 4: Code Generation")
print("-" * 50)
response = llm.invoke("Write a Python function to check if a number is prime")
print(response.content)
print("\n")

# Example 5: With custom parameters
print("Example 5: Custom Parameters (High Creativity)")
print("-" * 50)
creative_llm = ChatAnthropic(
    model="claude-sonnet-4-5-20250929",
    temperature=1.0,  # Maximum creativity
    max_tokens=200
)
response = creative_llm.invoke("Give me a creative name for a coffee shop")
print(response.content)
```
