```python
from langchain_openai import ChatOpenAI

# Initialize ChatGPT model
llm = ChatOpenAI(
    model="gpt-4o",  # You can also use "gpt-4", "gpt-3.5-turbo", etc.
    temperature=0.7,
    max_tokens=None,
    timeout=None,
    max_retries=2,
    # api_key="...",  # Optional: pass directly instead of using env var
)

# Simple chat

response = llm.invoke("Hello, how are you?")
print(response)
```
